```python
#we draw a histogarm to show the age appearence :
plt.figure(figsize=(20,10))#we createwith dimensions we want the paper where we draw 
plt.hist(df['Age'],bins=10,color="Purple",edgecolor='Black')#drawing a histo(i like purple)
plt.title("age appearance")#we add title to our histo
plt.show()
```

```python
#as we notice , the first class has the highest fare !
#now we want to see the sex of survivors!
plt.figure(figsize=(10,10))#we create paper with dimensions we want again to not overwrite

sex_sur = df.groupby('Sex')['Survived'].mean()
#we group the data frame depending on the sex , then we calculate the average of the survivors per a sex:
# Create the bar chart
plt.bar(sex_sur.index, sex_sur.values, color="Purple", edgecolor="Black")
#we put the sex in the x axis and the average to be in the y axis ,its like affecting each mean to the sex by following the indexing.
plt.title("survivors statistics")
plt.show()

```

```python
#removing the  unnecessary columns
df.drop(["Name","PassengerId","Pclass","SibSp","Parch","Ticket","Fare","Cabin","Embarked"],axis=1,inplace=True)
df
```

```python
# handling missing values:
df.isna()
```

```python
missing_counts = df.isna().sum()
print("Missing values per column:")
print(missing_counts)
```

```python
fill_na = {
    "Age": df["Age"].median(),       
}
df = df.fillna(fill_na)
missing_counts = df.isna().sum()
print("Missing values per column:")
print(missing_counts)
```

```python
#converting categorical data
df['Sex']=pd.factorize(df['Sex'])[0]
df
```
